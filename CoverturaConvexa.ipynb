# Introducción

La geometría computacional, una fascinante disciplina que surge de la intersección entre las matemáticas y la informática, se ocupa de explorar cuestiones fundamentales relacionadas con la forma, el tamaño, la posición relativa de las figuras y las propiedades del espacio. En el vasto universo de la informática, la geometría computacional se erige como una rama dedicada al estudio de algoritmos que pueden expresarse de manera geométrica, proporcionando herramientas y métodos para abordar problemas complejos.

¿Dónde encuentra aplicación concreta la geometría computacional en el mundo real? Podemos vislumbrar su presencia en diversos escenarios, como en la creación de entornos virtuales para videojuegos. Aquí, la representación precisa del entorno mientras el jugador se desplaza requiere cuidadosa consideración de factores como la dirección de la vista, el movimiento durante eventos específicos y otros aspectos que son gestionados con destreza mediante la geometría computacional.

En el ámbito de la geometría computacional, los problemas tridimensionales o de dimensiones superiores son escasos, principalmente debido a la complejidad de los algoritmos, que tiende a aumentar significativamente con el número de dimensiones involucradas. La mayoría de los desafíos se centran en la geometría euclidiana bidimensional, destacando la importancia de abordar problemas prácticos con soluciones eficientes y efectivas.

Al explorar los problemas relacionados con la geometría computacional, nos sumergimos en un intrigante mundo teórico que a menudo demanda pruebas detalladas e intuiciones perspicaces. Contrariamente a la noción común de que entender superficialmente algunos problemas geométricos puede ser suficiente, a largo plazo, la complejidad y diversidad de estos desafíos exigen un profundo conocimiento y comprensión de los fundamentos geométricos y algoritmos asociados.

Un ejemplo ilustrativo de la aplicación práctica de la geometría computacional es el "problema de la galería de arte". Imaginemos que somos propietarios de una galería de arte y deseamos instalar cámaras de seguridad de manera eficiente para proteger nuestras obras. Aquí, la notación geométrica computacional nos permite abordar este problema, demostrando que n/3 cámaras son siempre suficientes para un polígono con n vértices, independientemente de su complejidad. La resolución implica el uso ingenioso de gráficos duales, teoría de grafos, triangulaciones y más, resaltando la capacidad de la geometría computacional para proporcionar soluciones prácticas a desafíos del mundo real. En este sentido, se la geometría computacional se revela como una herramienta poderosa que va más allá de la mera abstracción teórica, ofreciendo resultados tangibles y aplicables en situaciones cotidianas.

**Visualizar elementos geométricos**

```python
# Crear un gráfico con Matplotlib
fig, ax = plt.subplots()

# Obtener las coordenadas x e y de los puntos
x_P1, y_P1 = P1.x, P1.y
x_P2, y_P2 = P2.x, P2.y
x_P3,y_P3=P3.x,P3.y

# Dibujar puntos
ax.plot(x_P1, y_P1, 'ro', label='Punto 1')
ax.plot(x_P2, y_P2, 'go', label='Punto 2')
ax.plot(x_P3, y_P3, 'bo', label='Punto 2')

# Dibujar segmentos
ax.plot([x_P1, x_P2], [y_P1, y_P2], label='Segmento P1P2')
ax.plot([x_P2, x_P3], [y_P2, y_P3], label='Segmento P2P3')
ax.plot([x_P3, x_P1], [y_P3, y_P1], label='Segmento P3P1')

# Dibujar líneas rectas
eq12 = L12.equation() #Obtener la ecuación de la línea
print("Ecuación de la recta P1P2: ",eq12,"= 0")
m=float(L12.slope)
if np.isfinite(m):
  x_vals = np.linspace(-10, 10, 100)
  y_vals = m * x_vals - m*x_P1+y_P1
  plt.plot(x_vals, y_vals, label='Recta P1P2')
else:
  y_vals = np.linspace(-10, 10, 100)
  x_vals = 0 * y_vals - 0*y_P1+x_P1
  plt.plot(x_vals, y_vals, label='Recta P1P2')

# Dibujar circunferencias
eqC1=C1.equation() #Ecuación de la circunferencia
print("Ecuación de la circunferencia C1: ",eqC1,"= 0")
r=C1.radius #Radio de la circunferencia
theta = np.linspace(0, 2*np.pi, 100) #Definir el ángulo theta
x_vals =x_P1+ r * np.cos(theta)
y_vals =y_P1+ r * np.sin(theta)
plt.plot(x_vals, y_vals, label=f'Circunferencia C1')

# Hacer que los ejes tengan la misma escala
plt.axis('equal')

# Configurar el gráfico
ax.axhline(0, color='black',linewidth=0.5)
ax.axvline(0, color='black',linewidth=0.5)
ax.grid(color = 'gray', linestyle = '--', linewidth = 0.5)

# Etiquetas y leyenda
ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')
ax.legend(loc='upper right')

# Mostrar el gráfico
plt.show()
```

**Funciones de trazado**

Ubicación de puntos

```python
def GPunto(P,clr):
  ax.plot(P.x, P.y,marker='o',markersize=5,color=clr)
```

Trazado de segmentos

```python
def GSegmento(P1,P2):
  ax.plot([P1.x, P2.x], [P1.y, P2.y])
```

Trazado de lineas

```python
def GLinea(L):
  P=L.arbitrary_point()
  x_P, y_P=P.x, P.y
  m=float(L.slope)
  if np.isfinite(m):
    x_vals = np.linspace(-10, 10, 100)
    y_vals = m * x_vals - m*x_P+y_P
    plt.plot(x_vals, y_vals)
  else:
    y_vals = np.linspace(-10, 10, 100)
    x_vals = 0 * y_vals - 0*y_P+x_P
    plt.plot(x_vals, y_vals)
```

Trazado de circunferencias

```python
def GCircunferencia(P,r):
  theta = np.linspace(0, 2*np.pi, 100) #Definir el ángulo theta
  x_vals =P.x+ r * np.cos(theta)
  y_vals =P.y+ r * np.sin(theta)
  plt.plot(x_vals, y_vals, label=f'Circunferencia C1')
```

```python
def orientacion(x:Point,y:Point,p:Point)->np.int32:
  z=Point(y[0]-x[0],y[1]-x[1])
  q=z.rotate(np.pi/2)
  d=Point(p[0]-x[0],p[1]-x[1])
  alfa=q.dot(d)
  if np.float32(alfa) > 0:
    return np.int32(1)
  elif np.float32(alfa) < 0:
    return np.int32(-1)
  else:
    return np.int32(0)
```

```python
def CH(P:list)->list:
  dcc=orientacion(P[0],P[1],P[2])
  poligono=[P[0],P[1]]
  for j in range(1,len(P)-2):
    pivot=dcc
    dcc=orientacion(P[j],P[j+1],P[j+2])
    if dcc==pivot:
      poligono.append(P[j+2])
    else:
      poligono.pop(-1)
      poligono.append(P[j+2])
  return poligono
```
