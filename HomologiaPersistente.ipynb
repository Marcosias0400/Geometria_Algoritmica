```python
def s2xs2(r:float,err:float=0,N:int=100)->list:
  B=[]
  for i in range(int(N/2)):
    t=rnd.random()*2*np.pi
    s=np.arccos(rnd.random()*2-1)
    u=rnd.randint(-1,1)
    B.append(np.array([(r+u*err)*np.sin(t)*np.cos(s)-r, (r+u*err)*np.cos(t)*np.sin(s), (r+u*err)*np.cos(s)]))
  for i in range(int(N/2)):
    t=rnd.random()*2*np.pi
    s=np.arccos(rnd.random()*2-1)
    u=rnd.randint(-1,1)
    B.append(np.array([(r+u*err)*np.sin(t)*np.cos(s)+r, (r+u*err)*np.cos(t)*np.sin(s), (r+u*err)*np.cos(s)]))
  return B
```

```python
K=s2xs2(1,0,10000)
C=np.array(K)
x=[];y=[];z=[]
for k in range(len(C)):
  x.append(C[k][0])
  y.append(C[k][1])
  z.append(C[k][2])
fig = plt.figure()
ax = fig.add_subplot(projection = '3d')
# Gráfico de dispersión 3D
ax.scatter(x, y, z)
# Equivalente a:
ax.scatter3D(x, y, z)
plt.show()
L=np.array([x,y,z])
```

```python
H=cx.Cloud(data=L)
H.plot()
```

```python
o=H.persistent_homology()
o.persistence_diagram()
o.bar_code()
```

```python
def t2xt2(r:float,R:float,err:float=0,N:int=100)->list:
  B=[]
  for i in range(int(N/2)):
    t=rnd.random()*2*np.pi
    s=np.arccos(rnd.random()*2-1)
    u=rnd.randint(-1,1)
    B.append(np.array([np.cos(t)*(R+r*np.cos(s))-R, np.sin(t)*(R+r*np.cos(s)), (r+u*err)*np.sin(s)]))
  for i in range(int(N/2)):
    t=rnd.random()*2*np.pi
    s=np.arccos(rnd.random()*2-1)
    u=rnd.randint(-1,1)
    B.append(np.array([np.cos(t)*(R+r*np.cos(s))+R,np.sin(t)*(R+r*np.cos(s)),(r+u*err)*np.sin(s)]))
  return B
```

```python
L=t2xt2(1,3,0,10000)
C=np.array(L)
x=[];y=[];z=[]
for k in range(len(C)):
  x.append(C[k][0])
  y.append(C[k][1])
  z.append(C[k][2])
fig = plt.figure()
ax = fig.add_subplot(projection = '3d')
# Gráfico de dispersión 3D
ax.scatter(x, y, z)
# Equivalente a:
ax.scatter3D(x, y, z)
plt.show()
M=np.array([x,y,z])
```

```python
O=cx.Cloud(data=M)
O.plot()
```

```python
f=O.persistent_homology()
f.persistence_diagram()
f.bar_code()
```
